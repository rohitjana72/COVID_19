```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_Set=pd.read_csv("C:/Users/Rohit Jana/OneDrive/Desktop/stats notes/covid19_Confirmed_dataset.csv")
data_Set.head()
```

```python
#size of the dataset
data_Set.shape  #266 rows and 104 columns
```

```python
#delete useless columns
data_Set.drop(["Lat","Long"],axis=1,inplace=True)
data_Set.head()
```

```python
#aggregate the rows by the country
aggregated_data_set=data_Set.groupby("Country/Region").sum() # to get summary of corresponding country/Region
aggregated_data_set
```

```python
#size of the aggregated data set
aggregated_data_set.shape   # 187 rows and 100 columns
```

```python
#visualize the data related to some countries
aggregated_data_set.loc["India"].plot()
aggregated_data_set.loc["China"].plot()
aggregated_data_set.loc["Spain"].plot()
plt.legend()
```

```python
#check covid in China in first 3 days
aggregated_data_set.loc["China"][:3].plot()
```

```python
# to calculate the rate of the curve
#calculate the first derivative of the curve
aggregated_data_set.loc["China"].diff().plot()
```

```python
#maximum infection rate
aggregated_data_set.loc["China"].diff().max()
```

```python
aggregated_data_set.loc["India"].diff().max()
```

```python
aggregated_data_set.loc["Spain"].diff().max()
```

```python
# calculate maximum infection rate for all countries
countries=list(aggregated_data_set.index)
max_infection_rates=[]

for c in countries:
    max_infection_rates.append(aggregated_data_set.loc[c].diff().max())
aggregated_data_set["Max_infection_rates"] = max_infection_rates
aggregated_data_set
```

```python
#create a new dataframe
corona_data=pd.DataFrame(aggregated_data_set['Max_infection_rates'])
corona_data
```

```python
corona_data['infection_rate %']=corona_data['Max_infection_rates'].apply(lambda x: 100 * x / corona_data['Max_infection_rates'].sum())
corona_data
```

```python
print(corona_data.max())
```

```python
max_ind = corona_data['infection_rate %'].idxmax() # Index of maximum in column
print(max_ind)
```
